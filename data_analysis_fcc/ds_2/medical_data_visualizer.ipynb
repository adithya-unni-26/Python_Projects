```python
#cleaning the dataset for the heatmap
```

```python
df=df.loc[(df['ap_lo']) <= (df['ap_hi'])]
```

```python
df=df.loc[(df['height']) >= (df['height'].quantile(0.025))]
```

```python
df=df.loc[~(df['height']>(df['height'].quantile(0.975)))]
```

```python
df=df.loc[~(df['weight']<(df['weight'].quantile(0.025)))]
```

```python
df=df.loc[~(df['weight']>(df['weight'].quantile(0.975)))]
```

```python
df.drop(columns='Overweight')
```

```python
corr=df.corr()
```

```python
mask=np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True
# mask[np.diag_indices_from(mask)] = False
```

```python
f, ax = plt.subplots(figsize=(20, 9))
```

```python
cmap = sns.color_palette("cubehelix", as_cmap=True)
```

```python
sns_plot = sns.heatmap(data=corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
